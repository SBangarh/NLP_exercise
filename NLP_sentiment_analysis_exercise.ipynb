**Task:** Clean the text data in the `'features'` array.

    - Remove all the special characters.
    - Remove all single characters.
    - Remove single characters from the start.
    - Substituting multiple spaces with single space.
    - Converting all text to lowercase.

```python
def remove_special_char(your_string):
    pattern = r'[^A-Za-z0-9 ]' #define pattern to look for
    cleaned_string = re.sub(pattern, '', your_string) #replace special chars with nothing
    return cleaned_string

def remove_single_char(your_string):
    my_array = your_string.split()
    new_string = []
    for word in my_array:
        if len(word) >= 2:
            new_string.append(word)
        else:
            pass
    return new_string

def remove_multi_space(your_string):
    my_string = ' '.join(your_string)
    pattern = r'\s+'
    new_string = re.sub(pattern, ' ', my_string)
    return new_string.split()

def lower_case(your_string):
    lower_string = []
    for word in your_string:
        lower_string.append(word.lower())
    return lower_string


processed_features = []

for sentence in features:

    # Remove all the special characters
    new_string = remove_special_char(sentence)
    
    # Remove all single characters
    my_string = remove_single_char(new_string)
    
    # Remove single characters from the start this is above too

    # Substituting multiple spaces with single space
    new_list = remove_multi_space(my_string)

    # Converting to Lowercase
    lower_list = lower_case(new_list)
    
    #append to processed_features
    processed_features.append(' '.join(lower_list))

processed_features


    
```

**Task:** Import stopwords from nltk.

```python
from nltk.corpus import stopwords

def remove_stopwords(data):
    stop_words = stopwords.words('english') #make sure stopwords is downloaded from nltk
    filtered_data = [word for word in data if word not in stop_words]
    return filtered_data

clean_list = []
for sentence in processed_features:
    words = sentence.split()
    clean_list.append(remove_stopwords(words))

clean_list
```

**Task:** Import TfidfVectorizer from sklearn.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

**Task:** Instatiate TfidfVectorizer with following parameters:

**Task:** Transform features with vectorizer. 

```python
vec_data = vectorizer.fit_transform(processed_features)
```

**Task:** Import train_test_split from sklearn and split the data.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(vec_data, labels, shuffle=True)
```

**Task:** Import any classifier of your choice from sklearn (e.g. Random Forest, LogReg, Naive Bayes).

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

```

**Task:** Fit your classifier to data.

```python
rf.fit(X_train, y_train)
```

**Task:** Predict X_test.

```python
y_predict = rf.predict(X_test)
```
